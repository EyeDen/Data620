I took the infrastructure network from http://networkrepository.com/inf-USAir97.php.  I'm not quite sure what it's about to be honest.

The file had some attribution for the first 23 lines, but I deleted them to make opening the file and reading in the graph easier.

```python
import networkx as nx

fh = open("inf-USAir97/inf-USAir97.mtx", 'rb')
G = nx.read_adjlist(fh)
nx.draw_networkx(G)
```

```python
diam = nx.diameter(G)
den = nx.density(G)

edges = nx.number_of_edges(G)
nodes = nx.number_of_nodes(G)
cluster = nx.average_clustering(G)

print("Diameter: ", diam)
print("Density: ", den)
print("Number of edges: ", edges)
print("Average number of nodes: ", nodes)
print("Average clustering: ", cluster)
```
